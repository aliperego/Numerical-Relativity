## TOV Evolution

```python
import matplotlib.pyplot as plt
import numpy as np
import kuibit.cactus_scalars as cs
import kuibit.simdir as sd
from kuibit import grid_data as gd
```

### No physical perturbation (K=100)

```python
sim1 = sd.SimDir("../TOV_example/")
sim2 = sd.SimDir("../TOV_dx_15/")
sim3 = sd.SimDir("../TOV_dx_1/")
label= ['dx = 2.0', 'dx = 1.5','dx = 1.0']

plt.figure(figsize=(7,5))
for l,s in zip(label,[sim1,sim2,sim3]):
    timeseries = s.timeseries
    rho_max=timeseries.maximum.fields.rho
    plt.plot(rho_max, label=l)
plt.ylim(0.00122,0.00165)
plt.xlabel(r'$t$ [$M_{\odot}$]')
plt.ylabel(r'$\rho_c$')
plt.legend()
plt.savefig('TOV_fig2.pdf', dpi=200)
```
