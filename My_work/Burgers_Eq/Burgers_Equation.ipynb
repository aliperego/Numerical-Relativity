# Burgers' Equation

I solve the Burgers' equation $u_t + u u_x=0$ using as initial condition a gaussian profile $u(x,t=0)=10e^{-(x-x_0)^2}$ with $x_0=5$.

The domain is $x\in [0,10]$ and the final time is t = 0.5.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

### N grid points = 101

```python
L = 10.0
nx = 101
dx = L/(nx-1)
x = np.linspace(0, L, nx)

x0=5
u_initial = 10*np.exp(-(x-x0)**2)

cf = 0.5
dt = cf*dx/(np.max(u_initial))
t_final = 0.5

print('nx=',nx)
print('dx=',dx)
print('dt=',dt)
print('Number of iterations=',t_final/dt)
```

#### Conservative form

```python
os.makedirs('./Conservative/images_101', exist_ok=True)

u_current = u_initial.copy()
time=[]
time.append(0.0)
t = 0.0
i = 0

plt.plot(x, u_current)
plt.title('Time='+str(round(t,2)))
plt.ylim(-0.5,10.5)
plt.xlabel('x')
plt.ylabel('u', rotation=0)
plt.savefig('./Conservative/images_101/fig_'+str(i).zfill(5)+'.png', dpi=200)
plt.close()

while t < t_final:
    
    u_next = u_current - (dt/dx)*(0.5*(u_current**2)-0.5*(np.roll(u_current, 1)**2))
    u_current = u_next.copy()
    t += dt
    i += 1

    if (i%5==0):
        plt.plot(x, u_current)
        plt.title('Time='+str(round(t,2)))
        plt.ylim(-0.5,10.5)
        plt.xlabel('x')
        plt.ylabel('u', rotation=0)
        plt.savefig('./Conservative/images_101/fig_'+str(i).zfill(5)+'.png', dpi=200)
        plt.close()

plt.plot(x, u_initial, label='Initial')
plt.plot(x, u_current, label='Final')
plt.title('Time='+str(round(t,2)))
plt.ylim(-0.5,10.5)
plt.xlabel('x')
plt.ylabel('u', rotation=0)
plt.legend()
plt.show()

u_final_101=u_current

directory = "./Conservative/images_101/"
files = os.listdir(directory)
files=sorted(files)

import imageio.v2 as imageio
with imageio.get_writer('./Conservative/movie_101.mov', mode='I', fps=10) as writer:
    for file in files:
        image = imageio.imread('./Conservative/images_101/'+file)
        writer.append_data(image)
        
files=[]
```

```python
os.makedirs('./NonConservative/images_101', exist_ok=True)

u_current = u_initial.copy()
time=[]
time.append(0.0)
t = 0.0
i = 0

plt.plot(x, u_current)
plt.title('Time='+str(round(t,2)))
plt.ylim(-0.5,10.5)
plt.xlabel('x')
plt.ylabel('u', rotation=0)
plt.savefig('./NonConservative/images_101/fig_'+str(i).zfill(5)+'.png', dpi=200)
plt.close()

while t < t_final:
    
    u_next = u_current - (dt/dx)*u_current*(u_current - np.roll(u_current, 1))
    u_current = u_next.copy()
    t += dt
    i += 1

    if (i%5==0):
        plt.plot(x, u_current)
        plt.title('Time='+str(round(t,2)))
        plt.ylim(-0.5,10.5)
        plt.xlabel('x')
        plt.ylabel('u', rotation=0)
        plt.savefig('./NonConservative/images_101/fig_'+str(i).zfill(5)+'.png', dpi=200)
        plt.close()
        
plt.plot(x, u_initial, label='Initial')
plt.plot(x, u_current, label='Final')
plt.title('Time='+str(round(t,2)))
plt.ylim(-0.5,10.5)
plt.xlabel('x')
plt.ylabel('u', rotation=0)
plt.legend()
plt.show()

u_final_101_non=u_current

directory = "./NonConservative/images_101/"
files = os.listdir(directory)
files=sorted(files)

import imageio.v2 as imageio
with imageio.get_writer('./NonConservative/movie_101.mov', mode='I', fps=10) as writer:
    for file in files:
        image = imageio.imread('./NonConservative/images_101/'+file)
        writer.append_data(image)
        
files=[]
```
